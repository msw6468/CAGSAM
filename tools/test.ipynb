```python
import pandas as pd
import os, sys
from glob import glob
import shutil
df = pd.read_csv('FineARCADE_info.csv')
df = df[['group', 'png_name', 'new_png_num', 'new_png_name']]
df['group'][df['group'] == 'r1'] = 'test'
df['group'][df['group'] != 'test'] = 'train'
df['image_filename'] = 'xxx'
df['mask_filename'] = 'xxx'
df['main_mask_filename'] = 'xxx'
df['fine_mask_filename'] = 'xxx'
for i in range(len(df)):
    split = df['group'][i]
    original_idx = int(df['png_name'][i].split('.')[0])
    df['image_filename'][i] = f'data/ARCADE/syntax/{split}/images/{original_idx}.png'
    df['mask_filename'][i] = f'data/FineARCADE/syntax/{split}/masks/{original_idx}.png'
    df['main_mask_filename'][i] = f'data/FineARCADE/syntax/{split}/main_masks/{original_idx}.png'
    df['fine_mask_filename'][i] = f'data/FineARCADE/syntax/{split}/fine_masks/{original_idx}.png'

    # os.makedirs(os.path.dirname(df['mask_filename'][i]), exist_ok=True)
    # os.makedirs(os.path.dirname(df['main_mask_filename'][i]), exist_ok=True)
    # os.makedirs(os.path.dirname(df['fine_mask_filename'][i]), exist_ok=True)

    # shutil.move(f'data/FineARCADE/RCA_masks_annotations/masks/{df["new_png_name"][i]}', df['mask_filename'][i])
    # shutil.move(f'data/FineARCADE/RCA_masks_annotations/main_masks/{df["new_png_name"][i]}', df['main_mask_filename'][i])
    # shutil.move(f'data/FineARCADE/RCA_masks_annotations/rca_masks/{df["new_png_name"][i]}', df['fine_mask_filename'][i])

df['origin_split'] = df['group']
df['split'] = 'test'
df = df[['split', 'image_filename', 'mask_filename', 'main_mask_filename', 'fine_mask_filename']]

for i in range(len(df)):
    if not os.path.exists(f'{df.iloc[i]["image_filename"]}'):
        print(os.path.exists(f'{df.iloc[i]["image_filename"]}'))

filelist = glob('data/ARCADE/syntax/*/images/*.png')
filelist.sort()

for filename in filelist:
    if filename not in list(df['image_filename']):
        assert os.path.exists(filename.replace('images','masks'))
        df = pd.concat([df, pd.DataFrame({
            'image_filename': [filename],
            'mask_filename' : [filename.replace('images','masks')],
            'split'         : ['train']})])
df.to_csv('data/FineARCADE/data_info.csv')
df2 = pd.read_csv('data/FineARCADE/data_info.csv', index_col=0)
df2
```
